```python
# STEP 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score

# STEP 2: Load dataset
df = pd.read_csv("/content/single_genre_artists (10).csv")

# STEP 3: Select relevant numeric audio features
features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]

df_features = df[features]

# STEP 4: Scale the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled = pd.DataFrame(df_scaled, columns=features)

# STEP 5: Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(df_scaled)

# STEP 6: Add cluster labels
df_clustered = df_scaled.copy()
df_clustered['Cluster'] = dbscan_labels

# STEP 7: Visualize DBSCAN Clusters (first 2 features or PCA later)
plt.figure(figsize=(8,6))
plt.scatter(df_clustered.iloc[:, 0], df_clustered.iloc[:, 1],
            c=dbscan_labels, cmap='plasma', alpha=0.6)
plt.title('DBSCAN Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# STEP 8: Evaluate Clustering
if len(set(dbscan_labels)) > 1:
    sil = silhouette_score(df_scaled, dbscan_labels)
    dbi = davies_bouldin_score(df_scaled, dbscan_labels)
    print(f"Silhouette Score: {sil:.3f}")
    print(f"Davies-Bouldin Index: {dbi:.3f}")
else:
    print("Not enough clusters for evaluation metrics.")

# STEP 9: Inspect Cluster Counts
n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print(f"\nNumber of clusters (excluding noise): {n_clusters}")
print(f"Number of noise points: {(dbscan_labels == -1).sum()}")
```

```python
# STEP 1: Import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

# STEP 2: Load your dataset
df = pd.read_csv("/content/single_genre_artists (10).csv")

# STEP 3: Select numerical features for clustering
features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]
df_features = df[features]

# STEP 4: Scale the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled = pd.DataFrame(df_scaled, columns=features)

# STEP 5: Plot dendrogram to visualize cluster structure
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(df_scaled, method='ward'))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Songs')
plt.ylabel('Euclidean Distance')
plt.show()

# STEP 6: Perform Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=5, linkage='ward')
agg_labels = agg.fit_predict(df_scaled)

# STEP 7: Add cluster labels to DataFrame
df_hierarchical = df_scaled.copy()
df_hierarchical['Cluster'] = agg_labels

# STEP 8: Visualize the clusters (first two features)
plt.figure(figsize=(8,6))
plt.scatter(df_hierarchical.iloc[:, 0], df_hierarchical.iloc[:, 1],
            c=agg_labels, cmap='viridis', alpha=0.6)
plt.title('Hierarchical Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

print("✅ Number of clusters formed:", len(set(agg_labels)))
```

```python
# ==========================================
# 🎵 SONG CLUSTERING USING K-MEANS
# ==========================================

# STEP 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

# STEP 2: Load dataset
df = pd.read_csv("/content/single_genre_artists (10).csv")

# STEP 3: Select audio features useful for clustering
features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]
df_features = df[features]

# STEP 4: Scale features for clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled = pd.DataFrame(df_scaled, columns=features)

# ==========================================
# STEP 5: Find optimal number of clusters
# Using the Elbow Method
# ==========================================
inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method: Optimal Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.grid(True)
plt.show()

# ==========================================
# STEP 6: Evaluate using Silhouette Scores
# ==========================================
print("\nSilhouette Scores for Different k:")
for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_scaled)
    sil_score = silhouette_score(df_scaled, labels)
    print(f"k = {k} → Silhouette Score = {sil_score:.3f}")

# ==========================================
# STEP 7: Apply Final K-Means Model
# (Choose k based on Elbow & Silhouette)
# ==========================================
optimal_k = 5   # 🔹 You can adjust this after reviewing the Elbow plot

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(df_scaled)

# Add cluster labels to the dataset
df_kmeans = df.copy()
df_kmeans['Cluster'] = kmeans_labels

# ==========================================
# STEP 8: Visualize Clusters using PCA
# ==========================================
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = kmeans_labels

plt.figure(figsize=(8,6))
plt.scatter(df_pca['PCA1'], df_pca['PCA2'],
            c=df_pca['Cluster'], cmap='rainbow', alpha=0.7)
plt.title(f'K-Means Clustering Visualization (k={optimal_k})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# ==========================================
# STEP 9: Evaluate the Final Model
# ==========================================
silhouette_avg = silhouette_score(df_scaled, kmeans_labels)
print(f"\n✅ Final Model Evaluation:")
print(f"Optimal K: {optimal_k}")
print(f"Silhouette Score: {silhouette_avg:.3f}")
print(f"Inertia (SSE): {kmeans.inertia_:.2f}")

# ==========================================
# STEP 10: Interpret Clusters
# ==========================================
cluster_summary = df_kmeans.groupby('Cluster')[features].mean().round(2)
print("\n🎧 Cluster Feature Means (for interpretation):\n")
print(cluster_summary)

```

```python
# ==========================================
# 🎧 SONG CLUSTERING: K-MEANS + EVALUATION
# ==========================================

# STEP 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA

# STEP 2: Load dataset
df = pd.read_csv("/content/single_genre_artists (10).csv")

# STEP 3: Select relevant audio features
features = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness',
    'valence', 'tempo', 'duration_ms'
]
df_features = df[features]

# STEP 4: Scale features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
df_scaled = pd.DataFrame(df_scaled, columns=features)

# STEP 5: Apply K-Means Clustering
optimal_k = 5  # you can change this after checking the Elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(df_scaled)

# Add cluster labels to the dataset
df_kmeans = df.copy()
df_kmeans['Cluster'] = kmeans_labels

# ==========================================
# STEP 6: Cluster Evaluation
# ==========================================

# Silhouette Score (higher is better)
silhouette_avg = silhouette_score(df_scaled, kmeans_labels)

# Davies-Bouldin Index (lower is better)
db_index = davies_bouldin_score(df_scaled, kmeans_labels)

# Inertia (compactness of clusters)
inertia_value = kmeans.inertia_

print("🔍 CLUSTER EVALUATION METRICS")
print(f"Silhouette Score: {silhouette_avg:.3f}  (higher = better)")
print(f"Davies-Bouldin Index: {db_index:.3f}  (lower = better)")
print(f"Inertia (SSE): {inertia_value:.2f}")

# ==========================================
# STEP 7: Cluster Interpretation
# ==========================================

cluster_profile = df_kmeans.groupby('Cluster')[features].mean().round(2)
print("\n🎧 CLUSTER PROFILES (Mean Feature Values per Cluster):\n")
print(cluster_profile)

# Plot feature means per cluster
cluster_profile.T.plot(kind='bar', figsize=(10,6))
plt.title('Average Feature Values per Cluster')
plt.xlabel('Audio Features')
plt.ylabel('Mean (Scaled Value)')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

# ==========================================
# STEP 8: Visualize Clusters in 2D (PCA)
# ==========================================

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(pca_result, columns=['PCA1', 'PCA2'])
df_pca['Cluster'] = kmeans_labels

plt.figure(figsize=(8,6))
plt.scatter(df_pca['PCA1'], df_pca['PCA2'],
            c=df_pca['Cluster'], cmap='rainbow', alpha=0.7)
plt.title(f'K-Means Clustering Visualization (k={optimal_k})')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# ==========================================
# STEP 9: Example Manual Interpretations
# ==========================================

print("""
💡 INTERPRETATION GUIDE:
- Cluster 0 → High danceability & energy → 'Party / EDM Tracks'
- Cluster 1 → High acousticness & low loudness → 'Chill Acoustic / Coffeehouse'
- Cluster 2 → High speechiness → 'Rap / Spoken Word'
- Cluster 3 → High valence (positive mood) → 'Happy Pop Songs'
- Cluster 4 → High instrumentalness → 'Instrumental / Ambient'
""")
```
